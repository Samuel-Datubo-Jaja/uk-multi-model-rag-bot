```python
# ======================= CORE RAG SYSTEM IMPLEMENTATION =======================
import os
import re
import time
import json
import concurrent.futures
from typing import List, Dict, Union, Optional, Callable, Any
from pathlib import Path
from abc import ABC, abstractmethod

import numpy as np
import pandas as pd
from tqdm import tqdm
from dotenv import load_dotenv

# LangChain components
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_chroma import Chroma
from langchain_groq import ChatGroq
from gradio_client import Client

# Search optimization
from rank_bm25 import BM25Okapi

```

```python
# ======================= BASE RAG SYSTEM ABSTRACT CLASS =======================

class BaseRAG(ABC):
    """Abstract base class for all RAG implementations with common functionality"""
    
    def __init__(self, model_name: str, vector_store_dir: str = "./main_chroma_data"):
        """Initialize the base RAG system"""
        self.model_name = model_name
        
        # Initialize embeddings - shared across all implementations
        try:
            self.embeddings = HuggingFaceEmbeddings(
                model_name="sentence-transformers/all-mpnet-base-v2",
                encode_kwargs={'normalize_embeddings': True}
            )
            
            # Initialize vector store
            self.vectorstore = Chroma(
                collection_name="main_construction_rag",
                embedding_function=self.embeddings,
                persist_directory=vector_store_dir
            )
        except Exception as e:
            print(f"Error initializing embeddings or vectorstore: {str(e)}")
            raise
            
        # Track metrics
        self.metrics = {
            "retrieval_times": [],
            "generation_times": [],
            "total_times": []
        }
    
    def hybrid_search(self, query: str, k: int = 4) -> List[str]:
        """Combine vector search with BM25 keyword search for better results"""
        # Get documents from vector search
        retrieval_start = time.time()
        vector_docs = self.vectorstore.similarity_search(query, k=k*2)
        
        # Extract content and metadata
        contents = [doc.page_content for doc in vector_docs]
        
        # Prepare corpus for BM25
        tokenized_corpus = [doc.split() for doc in contents]
        bm25 = BM25Okapi(tokenized_corpus)
        
        # Get BM25 scores
        tokenized_query = query.split()
        bm25_scores = bm25.get_scores(tokenized_query)
        
        # Normalize scores to 0-1 range
        if np.max(bm25_scores) > 0:
            bm25_scores = bm25_scores / np.max(bm25_scores)
        
        # Combine with vector search position (proxy for vector similarity)
        # Weight: 70% vector similarity, 30% BM25
        vector_weights = np.linspace(1, 0, len(contents))
        combined_scores = (0.7 * vector_weights) + (0.3 * bm25_scores)
        
        # Get top k results based on combined score
        top_indices = np.argsort(combined_scores)[::-1][:k]
        top_docs = [vector_docs[i] for i in top_indices]
        
        retrieval_time = time.time() - retrieval_start
        self.metrics["retrieval_times"].append(retrieval_time)
        
        return [doc.page_content for doc in top_docs]

    def get_most_relevant_docs(self, query: str, k: int = 4) -> List[str]:
        """Get relevant documents using hybrid search"""
        return self.hybrid_search(query, k=k)
    
    @abstractmethod
    def generate_answer(self, query: str, context: List[str]) -> str:
        """Generate answer based on query and context - to be implemented by subclasses"""
        pass
    
    def clean_response(self, response: str) -> str:
        """Clean up model response by removing instruction tags and truncating repetition"""
        # First, ensure response is a string
        if not isinstance(response, str):
            # Handle non-string response
            if hasattr(response, '__str__'):
                response = str(response)
            else:
                return "Error: Received non-string response from model"
        
        # Continue with normal cleaning
        cleaned = response.replace("[INST]", "").replace("[/INST]", "")
        
        # Remove any repeated paragraphs (a simple approach)
        paragraphs = []
        seen = set()
        for paragraph in cleaned.split("\n\n"):
            # Only keep paragraph if we haven't seen it before
            if paragraph.strip() and paragraph.strip() not in seen:
                paragraphs.append(paragraph)
                seen.add(paragraph.strip())
        
        return "\n\n".join(paragraphs)   
    
    
    
    def process_query(self, query: str, k: int = 4) -> Dict[str, Any]:
        """Process a query end-to-end and return results with timing information"""
        result = {}
        
        # Get relevant documents
        retrieval_start = time.time()
        contexts = self.get_most_relevant_docs(query, k=k)
        retrieval_time = time.time() - retrieval_start
        
        # Generate answer
        generation_start = time.time()
        response = self.generate_answer(query, contexts)
        generation_time = time.time() - generation_start
        
        # Calculate total time
        total_time = retrieval_time + generation_time
        
        # Update metrics
        self.metrics["retrieval_times"].append(retrieval_time)
        self.metrics["generation_times"].append(generation_time)
        self.metrics["total_times"].append(total_time)
        
        # Compile result
        result = {
            "query": query,
            "contexts": contexts,
            "response": response,
            "retrieval_time": retrieval_time,
            "generation_time": generation_time,
            "total_time": total_time
        }
        
        return result
    
    def get_metrics_summary(self) -> Dict[str, float]:
        """Get summary statistics for the performance metrics"""
        summary = {}
        
        if self.metrics["retrieval_times"]:
            summary["avg_retrieval_time"] = np.mean(self.metrics["retrieval_times"])
            summary["avg_generation_time"] = np.mean(self.metrics["generation_times"])
            summary["avg_total_time"] = np.mean(self.metrics["total_times"])
            summary["max_total_time"] = np.max(self.metrics["total_times"])
            summary["min_total_time"] = np.min(self.metrics["total_times"])
        
        return summary
```

```python
# ======================= CONCRETE RAG IMPLEMENTATIONS =======================

class GroqRAG(BaseRAG):
    """RAG implementation using GROQ API for LLM inference"""
    
    def __init__(
        self, 
        model_name: str = "llama-3.3-70b-versatile", 
        temperature: float = 0.1,
        vector_store_dir: str = "./main_chroma_data"
    ):
        """Initialize GROQ-based RAG system"""
        # Call parent constructor
        super().__init__(model_name, vector_store_dir)
        
        # Get API key
        groq_api_key = os.getenv("GROQ_API_KEY")
        if not groq_api_key:
            raise ValueError("GROQ_API_KEY not found in environment variables")
        
        # Initialize LLM
        try:
            self.llm = ChatGroq(
                api_key=groq_api_key,
                model_name=model_name,
                temperature=temperature
            )
            
            # Set descriptive name
            self.descriptive_name = f"GROQ {model_name} (temp={temperature})"
            
        except Exception as e:
            print(f"Error initializing GROQ LLM: {str(e)}")
            raise
    
    def generate_answer(self, query: str, context: List[str]) -> str:
        """Generate answer using GROQ"""
        combined_context = "\n\n".join(context)
        prompt = f"""Based on the following context from UK Building Regulations, provide a clear and detailed answer to the question.
        Include specific references to regulations where available.
        
        Question: {query}
        
        Context: {combined_context}
        
        Answer:"""
        
        try:
            generation_start = time.time()
            response = self.llm.invoke(prompt)
            generation_time = time.time() - generation_start
            
            self.metrics["generation_times"].append(generation_time)
            
            # Return cleaned response
            return self.clean_response(response.content)
            
        except Exception as e:
            print(f"Error generating answer with GROQ: {str(e)}")
            raise
```

```python
class HuggingFaceRAG(BaseRAG):
    """RAG implementation using Hugging Face hosted model (fine-tuned)"""
    
    def __init__(
        self, 
        hf_space_url: str = "https://samueljaja-llama3-1-8b-merged-uk-building-regulations.hf.space", #using the deployed model here frm Huggingface
        temperature: float = 0.1,
        max_tokens: int = 256,
        top_p: float = 0.9,
        vector_store_dir: str = "./main_chroma_data"
    ):
        """Initialize Hugging Face-based RAG system"""
        # Set model name for base class
        super().__init__("Fine-tuned Llama 3.1-8B", vector_store_dir)
        
        # Store parameters
        self.temperature = temperature
        self.max_tokens = max_tokens
        self.top_p = top_p
        self.descriptive_name = f"Fine-tuned Llama 3.1-8B (HF, temp={temperature})"
        
        # Initialize Gradio client
        try:
            self.client = Client(hf_space_url)
        except Exception as e:
            print(f"Error connecting to Hugging Face space: {str(e)}")
            raise
    
    def generate_answer(
        self, 
        query: str, 
        context: List[str],
        temperature: Optional[float] = None,
        max_tokens: Optional[int] = None,
        top_p: Optional[float] = None
    ) -> str:
        """Generate answer using Hugging Face hosted model"""
        # Use provided values or defaults
        temperature = temperature if temperature is not None else self.temperature
        max_tokens = max_tokens if max_tokens is not None else self.max_tokens
        top_p = top_p if top_p is not None else self.top_p
        
        combined_context = "\n\n".join(context)
        prompt = f"""Based on the following context from UK Building Regulations, provide a clear and detailed answer to the question.
        Include specific references to regulations where available.

        Question: {query}

        Context: {combined_context}

        Instructions:
        # 1. Answer directly and concisely
        # 2. Cite specific regulations and document sections when applicable
        # 3. Format measurements and requirements clearly
        # 4. If the exact information is not in the context, state this clearly
        # 5. For technical specifications, use proper formatting
        # 6. If you need to show a table, format it as markdown

        Answer:"""

        # Try up to 5 times with exponential backoff
        max_retries = 5
        for attempt in range(max_retries):
            try:
                print(f"Attempt {attempt+1}/{max_retries} to get response from Hugging Face model...")
                
                generation_start = time.time()
                result = self.client.predict(
                    prompt,          # prompt
                    temperature,     # temperature
                    max_tokens,      # max_tokens
                    top_p,           # top_p
                    api_name="/timed_predict_1"
                )
                generation_time = time.time() - generation_start
                
                self.metrics["generation_times"].append(generation_time)

                # Check if result is a tuple and extract the first element
                if isinstance(result, tuple) and len(result) > 0:
                    result = result[0]  # Extract just the response text
                # Additional post-processing to handle follow-up questions
                if result:
                    # Trim at any follow-up questions
                    question_patterns = [
                        r'\s+What\s+is', r'\s+What\s+are', r'\s+How\s+does', r'\s+How\s+can',
                        r'\s+Why\s+is', r'\s+Why\s+are', r'\s+When\s+should', r'\s+Where\s+can',
                        r'\?[\s\n]+[A-Z]',  # This catches questions ending with ? followed by new sentence
                        r'</INST>.*'  # This will trim anything after a closing instruction tag
                    ]
                    for pattern in question_patterns:
                        parts = re.split(pattern, result, 1)
                        if len(parts) > 1:
                            result = parts[0].strip()
                
                # Return cleaned response
                return self.clean_response(result)

            except Exception as e:
                # Check for timeout or connection errors
                if "timeout" in str(e).lower() or "connection" in str(e).lower():
                    # Wait longer for timeout errors
                    wait_time = 30 + (2 ** attempt) 
                    print(f"Connection timed out. Waiting {wait_time} seconds before retry...")
                    time.sleep(wait_time)
                    if attempt == max_retries - 1:
                        raise Exception(f"Failed to connect after {max_retries} attempts: {str(e)}")
                else:
                    # For other errors, use standard backoff
                    if attempt < max_retries - 1:
                        wait_time = 2 ** attempt
                        print(f"Error: {str(e)}. Retrying in {wait_time} seconds...")
                        time.sleep(wait_time)
                    else:
                        raise Exception(f"Failed after {max_retries} attempts: {str(e)}")
```

```python
# ======================= TEST CASE GENERATION =======================

class TestCaseGenerator:
    """Generates test cases for RAG evaluation using Giskard"""
    
    def __init__(self, vectorstore, openai_api_key: str, max_documents: int = 1000):
        """Initialize test generator with vectorstore"""
        self.vectorstore = vectorstore
        self.openai_api_key = openai_api_key
        self.max_documents = max_documents
        
        # Load knowledge base
        print("Loading documents from vectorstore...")
        self._load_documents()
    
    def _load_documents(self):
        """Load documents from vectorstore and initialize knowledge base"""
        try:
            # Get documents using similarity search
            results = self.vectorstore.similarity_search(
                "building regulations",
                k=self.max_documents
            )
            
            # Create DataFrame from documents
            documents = [{'text': doc.page_content} for doc in results]
            df = pd.DataFrame(documents)
            
            print(f"Loaded {len(df)} documents")
            
            if self._verify_documents(df):
                # Import here to avoid early dependency errors
                from giskard.rag import KnowledgeBase
                self.knowledge_base = KnowledgeBase(df)
                self.documents_df = df
            else:
                raise ValueError("Document verification failed")
                
        except Exception as e:
            print(f"Error loading documents via standard method: {str(e)}")
            self._try_alternative_loading()
    
    def _try_alternative_loading(self):
        """Try alternative method to load documents"""
        try:
            print("\nTrying alternative document loading method...")
            collection = self.vectorstore._collection
            docs = collection.get()
            documents = [{'text': text} for text in docs['documents']]
            df = pd.DataFrame(documents)
            
            if self._verify_documents(df):
                from giskard.rag import KnowledgeBase
                self.knowledge_base = KnowledgeBase(df)
                self.documents_df = df
            else:
                raise ValueError("Document verification failed")
                
        except Exception as e:
            print(f"Error in alternative document loading method: {str(e)}")
            raise
    
    def _verify_documents(self, df: pd.DataFrame) -> bool:
        """Verify that documents are loaded correctly"""
        if len(df) == 0:
            print("No documents loaded!")
            return False
        
        print(f"\nDocument Verification:")
        print(f"Total documents: {len(df)}")
        print("\nFirst document preview:")
        print(df['text'].iloc[0][:200] + "...")
        
        empty_docs = df['text'].isna().sum()
        if empty_docs > 0:
            print(f"\nWarning: Found {empty_docs} empty documents")
        
        return len(df) > 0 and empty_docs == 0
    
    def generate_test_cases(self, num_questions: int = 10) -> pd.DataFrame:
        """Generate test cases using Giskard"""
        print("\nGenerating test cases...")
        
        # Set OpenAI API key in environment
        os.environ["OPENAI_API_KEY"] = self.openai_api_key
        
        # Import here to avoid early dependency issues
        from giskard.rag import generate_testset
        
        try:
            testset = generate_testset(
                knowledge_base=self.knowledge_base,
                num_questions=num_questions,
                agent_description="""A chatbot answering technical questions about UK Building Regulations,
                focusing on specific measurements, requirements, and specifications for construction."""
            )
            
            # Convert to pandas DataFrame
            test_set_df = testset.to_pandas()
            
            # Display generated questions
            print("\nGenerated Test Questions:")
            print("-" * 50)
            for idx, row in enumerate(test_set_df.iterrows(), 1):
                print(f"\nQuestion {idx}:")
                print(f"Q: {row[1]['question']}")
                print(f"A: {row[1]['reference_answer']}")
                print(f"Context: {row[1]['reference_context'][:200]}...")
                print("-" * 50)
            
            return test_set_df
            
        except Exception as e:
            print(f"Error generating test cases: {str(e)}")
            raise
    
    def save_test_cases(self, test_set_df: pd.DataFrame, output_dir: str = "evaluation_results") -> tuple:
        """Save test cases to CSV and JSON files"""
        output_path = Path(output_dir)
        output_path.mkdir(exist_ok=True)
        
        # Save in both formats
        csv_path = output_path / "test_cases.csv"
        json_path = output_path / "test_cases.json"
        
        test_set_df.to_csv(csv_path, index=False)
        test_set_df.to_json(json_path, orient='records', indent=2)
        
        print(f"\nTest cases saved to:")
        print(f"CSV: {csv_path}")
        print(f"JSON: {json_path}")
        
        return csv_path, json_path
```

```python
# ======================= UNIFIED EVALUATION FRAMEWORK =======================

class RAGEvaluator:
    """Unified framework for evaluating and comparing RAG systems"""
    
    def __init__(self, output_dir: str = "rag_evaluation_results"):
        """Initialize evaluator with output directory"""
        self.output_dir = Path(output_dir)
        self.output_dir.mkdir(exist_ok=True)
        self.results = {}
    
    def load_test_cases(self, file_path: str) -> pd.DataFrame:
        """Load test cases from CSV or JSON file"""
        file_path = Path(file_path)
        
        if not file_path.exists():
            raise FileNotFoundError(f"Test case file not found: {file_path}")
        
        if file_path.suffix.lower() == '.csv':
            test_cases = pd.read_csv(file_path)
        elif file_path.suffix.lower() == '.json':
            test_cases = pd.read_json(file_path, orient='records')
        else:
            raise ValueError(f"Unsupported file format: {file_path.suffix}")
        
        print(f"Loaded {len(test_cases)} test cases from {file_path}")
        return test_cases

    def evaluate_system(
        self, 
        rag_system: BaseRAG, 
        test_cases: pd.DataFrame,
        context_count: int = 4, 
        use_parallel: bool = False,
        max_workers: int = 4
    ) -> pd.DataFrame:
        """Evaluate a RAG system on the provided test cases"""
        results = []
        system_name = rag_system.descriptive_name
        
        print(f"\nEvaluating {system_name}...")
        
        # Function for processing a single test case
        def process_test_case(idx, row_data):
            """Process a single test case"""
            # Ensure row_data is a Series, not a tuple or string
            if isinstance(row_data, tuple):
                # If this is from iterrows(), get the second element
                row_data = row_data[1]
            
            # Access data with dictionary-style access
            question = row_data['question']
            reference_answer = row_data['reference_answer']
            reference_context = row_data['reference_context']
            
            try:
                # Get RAG system's response with timing
                retrieval_start = time.time()
                retrieved_context = rag_system.get_most_relevant_docs(question, k=context_count)
                retrieval_time = time.time() - retrieval_start
                
                generation_start = time.time()
                rag_answer = rag_system.generate_answer(question, retrieved_context)
                generation_time = time.time() - generation_start
                
                total_time = retrieval_time + generation_time
                
                result = {
                    'question': question,
                    'reference_answer': reference_answer,
                    'rag_answer': rag_answer,
                    'reference_context': reference_context,
                    'retrieved_context': '\n'.join(retrieved_context) if isinstance(retrieved_context, list) else retrieved_context,
                    'retrieval_time': retrieval_time,
                    'generation_time': generation_time,
                    'total_time': total_time,
                    'system': system_name
                }
                
                # Print progress
                print(f"\nProcessed question {idx}:")
                print(f"Q: {question}")
                print(f"Response time: {total_time:.2f}s (retrieval: {retrieval_time:.2f}s, generation: {generation_time:.2f}s)")
                
                return result
            except Exception as e:
                print(f"Error processing question {idx}: {str(e)}")
                # Return partial result with error information
                return {
                    'question': question,
                    'reference_answer': reference_answer,
                    'reference_context': reference_context,
                    'error': str(e),
                    'system': system_name,
                    'rag_answer': None  # Ensure this field exists even in error cases
                }
        
        #Processing test cases (either in parallel or sequentially)
        if use_parallel and len(test_cases) > 1:
            print(f"Using parallel processing with {max_workers} workers...")
            with concurrent.futures.ThreadPoolExecutor(max_workers=max_workers) as executor:
                # Create a list of futures
                futures = []
                for idx, row in enumerate(test_cases.iterrows(), 1):
                    # Pass the entire row tuple to the function
                    futures.append(executor.submit(process_test_case, idx, row))
                
                # Process as they complete
                for future in tqdm(concurrent.futures.as_completed(futures), total=len(futures)):
                    try:
                        result = future.result()
                        if result:  # Only add non-None results
                            results.append(result)
                    except Exception as e:
                        print(f"Future execution error: {str(e)}")
        else:
            # Sequential processing
            for idx, row in enumerate(tqdm(test_cases.iterrows(), total=len(test_cases)), 1):
                result = process_test_case(idx, row)
                if result:  # Only add non-None results
                    results.append(result)
        
        # Create empty dataframe if no results
        if not results:
            print("No valid results were returned!")
            # Create an empty DataFrame with expected columns
            results_df = pd.DataFrame(columns=[
                'question', 'reference_answer', 'rag_answer', 'reference_context', 
                'retrieved_context', 'retrieval_time', 'generation_time', 'total_time', 
                'system', 'error'
            ])
        else:
            # Convert to DataFrame
            results_df = pd.DataFrame(results)
        
        # Save system-specific results
        system_dir = self.output_dir / system_name.replace(" ", "_")
        system_dir.mkdir(exist_ok=True)
        results_df.to_csv(system_dir / "evaluation_results.csv", index=False)
        
        # Store in results dictionary
        self.results[system_name] = results_df
        
        # Generate summary for this system
        self._generate_system_summary(results_df, system_name)
        
        return results_df
    
    def _generate_system_summary(self, results_df: pd.DataFrame, system_name: str):
        """Generate a summary of evaluation results for a single system"""
        # Count total questions
        total_questions = len(results_df)
        
        # Count successful responses (no errors)
        successful = results_df[~results_df['rag_answer'].isna()].shape[0]
        
        # Calculate average response times if available
        if 'retrieval_time' in results_df.columns:
            avg_retrieval_time = results_df['retrieval_time'].mean()
            avg_generation_time = results_df['generation_time'].mean()
            avg_total_time = results_df['total_time'].mean()
            max_total_time = results_df['total_time'].max()
            min_total_time = results_df['total_time'].min()
        else:
            avg_retrieval_time = avg_generation_time = avg_total_time = max_total_time = min_total_time = "N/A"
        
        # Create summary
        summary = {
            'system': system_name,
            'total_questions': total_questions,
            'successful_responses': successful,
            'success_rate': successful / total_questions if total_questions > 0 else 0,
            'avg_retrieval_time': avg_retrieval_time,
            'avg_generation_time': avg_generation_time,
            'avg_total_time': avg_total_time,
            'max_total_time': max_total_time,
            'min_total_time': min_total_time
        }
        
        # Save system summary
        system_dir = self.output_dir / system_name.replace(" ", "_")
        system_dir.mkdir(exist_ok=True)
        pd.DataFrame([summary]).to_csv(system_dir / "summary.csv", index=False)
        
        # Print summary
        print(f"\nEvaluation Summary for {system_name}:")
        print(f"Total questions: {total_questions}")
        print(f"Successful responses: {successful} ({summary['success_rate']:.1%})")
        
        if avg_total_time != "N/A":
            print(f"Average response time: {avg_total_time:.2f}s (retrieval: {avg_retrieval_time:.2f}s, generation: {avg_generation_time:.2f}s)")
            print(f"Response time range: {min_total_time:.2f}s - {max_total_time:.2f}s")
    
    def generate_comparison_report(self):
        """Generate a comparative report of all evaluated systems"""
        if not self.results:
            print("No systems have been evaluated yet.")
            return
        
        print("\nGenerating comparison report...")
        
        # Compile summary data from all systems
        summaries = []
        for system_name in self.results.keys():
            system_dir = self.output_dir / system_name.replace(" ", "_")
            summary_path = system_dir / "summary.csv"
            
            if summary_path.exists():
                summary = pd.read_csv(summary_path)
                summaries.append(summary)
        
        # Combine summaries
        comparison_df = pd.concat(summaries)
        
        # Save comparison report
        comparison_df.to_csv(self.output_dir / "system_comparison.csv", index=False)
        
        # Generate visualization
        try:
            self._visualize_comparison(comparison_df)
        except Exception as e:
            print(f"Error generating visualization: {str(e)}")
        
        print(f"Comparison report saved to {self.output_dir / 'system_comparison.csv'}")
        
        return comparison_df
    
    def _visualize_comparison(self, comparison_df: pd.DataFrame):
        """Create visualizations for the comparison data"""
        import matplotlib.pyplot as plt
        
        # Ensure numeric values
        numeric_cols = ['success_rate', 'avg_total_time', 'avg_retrieval_time', 'avg_generation_time']
        for col in numeric_cols:
            if col in comparison_df.columns:
                comparison_df[col] = pd.to_numeric(comparison_df[col], errors='coerce')
        
        # Plot success rates
        if 'success_rate' in comparison_df.columns:
            plt.figure(figsize=(10, 6))
            plt.bar(comparison_df['system'], comparison_df['success_rate'])
            plt.title('Success Rate Comparison')
            plt.xlabel('System')
            plt.ylabel('Success Rate')
            plt.xticks(rotation=45, ha='right')
            plt.tight_layout()
            plt.savefig(self.output_dir / "success_rate_comparison.png")
            plt.close()
        
        # Plot response times
        if 'avg_total_time' in comparison_df.columns:
            plt.figure(figsize=(12, 8))
            
            # Create grouped bar chart for time components
            bar_width = 0.25
            index = np.arange(len(comparison_df))
            
            plt.bar(index, comparison_df['avg_retrieval_time'], bar_width, 
                   label='Retrieval Time')
            plt.bar(index + bar_width, comparison_df['avg_generation_time'], bar_width,
                   label='Generation Time')
            plt.bar(index + 2*bar_width, comparison_df['avg_total_time'], bar_width,
                   label='Total Time')
            
            plt.title('Response Time Comparison')
            plt.xlabel('System')
            plt.ylabel('Seconds')
            plt.xticks(index + bar_width, comparison_df['system'], rotation=45, ha='right')
            plt.legend()
            plt.tight_layout()
            plt.savefig(self.output_dir / "response_time_comparison.png")
            plt.close()
        
        print(f"Visualizations saved to {self.output_dir}")

    def run_giskard_evaluation(self, rag_system: BaseRAG, test_cases: pd.DataFrame):
        """Run Giskard evaluation on the RAG system and generate HTML report"""
        try:
            from giskard.rag import evaluate
            import matplotlib.pyplot as plt
            
            system_name = rag_system.descriptive_name
            system_dir = self.output_dir / system_name.replace(" ", "_")
            system_dir.mkdir(exist_ok=True)
            
            print(f"\nPreparing Giskard evaluation for {system_name}...")
            
            # Create a wrapper function for the RAG system
            def get_answer_fn(question, history=None):
                """Wrapper function for Giskard evaluation"""
                try:
                    # Get context and generate answer
                    context = rag_system.get_most_relevant_docs(question)
                    answer = rag_system.generate_answer(question, context)
                    return answer
                except Exception as e:
                    print(f"Error generating answer for question: {question}")
                    print(f"Error details: {str(e)}")
                    return ""
            
            # Create knowledge base
            from giskard.rag import KnowledgeBase
            documents = []
            for _, row in test_cases.iterrows():
                documents.append({'text': row['reference_context']})
            
            kb_df = pd.DataFrame(documents)
            knowledge_base = KnowledgeBase(kb_df)
            
            # Generate a test set from our knowledge base
            from giskard.rag import generate_testset
            
            print("Generating Giskard test set...")
            testset = generate_testset(
                knowledge_base=knowledge_base,
                num_questions=5,  # Small number for efficiency
                language="en",
                agent_description="A chatbot answering technical questions about UK Building Regulations."
            )
            
            # Save the testset
            testset_path = system_dir / "giskard_testset.jsonl"
            testset.save(str(testset_path))
            print(f"Saved Giskard test set to {testset_path}")
            
            # Run the evaluation
            print(f"\nRunning Giskard evaluation for {system_name}...")
            report = evaluate(
                get_answer_fn,
                testset=testset,
                knowledge_base=knowledge_base
            )
            
            # Save as HTML file
            html_path = system_dir / "giskard_evaluation_report.html"
            report.to_html(str(html_path))
            print(f"Generated HTML report saved to {html_path}")
            
            # Display HTML report in Jupyter notebook if available
            try:
                # Check if we're in a notebook environment
                from IPython import get_ipython
                from IPython.display import HTML, display, Markdown
                
                if get_ipython() is not None:
                    display(Markdown(f"## Giskard Evaluation Report for {system_name}"))
                    
                    # Read and display the HTML file
                    with open(html_path, 'r', encoding='utf-8') as f:
                        html_content = f.read()
                        
                    # Use an iframe for better isolation
                    iframe_html = f"""
                    <iframe 
                        srcdoc="{html_content.replace('"', '&quot;')}" 
                        width="100%" 
                        height="800px" 
                        style="border: 1px solid #ddd; border-radius: 5px;">
                    </iframe>
                    """
                    display(HTML(iframe_html))
                    print("Giskard report displayed in notebook")
            except ImportError:
                # Not in a notebook environment, continue silently
                pass
            except Exception as e:
                # If display fails, just continue with the evaluation
                print(f"Note: Could not display report in notebook: {str(e)}")
            
            # Generate metrics visualization if available
            if hasattr(report, 'metrics'):
                plt.figure(figsize=(12, 6))
                metrics = report.metrics
                plt.bar(metrics.keys(), metrics.values())
                plt.xticks(rotation=45, ha='right')
                plt.title(f'RAG Evaluation Metrics - {system_name}')
                plt.tight_layout()
                metrics_path = system_dir / "giskard_metrics_visualization.png"
                plt.savefig(metrics_path)
                plt.close()
                print(f"Saved metrics visualization to {metrics_path}")
                
                # Save metrics as JSON
                metrics_path = system_dir / "giskard_metrics.json"
                with open(metrics_path, 'w') as f:
                    json.dump(metrics, f, indent=2)
            
            print(f"\nGiskard evaluation complete for {system_name}!")
            return report
        
        except Exception as e:
            print(f"Error during Giskard evaluation: {str(e)}")
            import traceback
            print(traceback.format_exc())
            return None
    
    
    def run_ragas_evaluation(self, rag_system: BaseRAG, test_cases: pd.DataFrame):
        """Run RAGAS evaluation for comprehensive metrics"""
        try:
            system_name = rag_system.descriptive_name
            system_dir = self.output_dir / system_name.replace(" ", "_")
            system_dir.mkdir(exist_ok=True)
            
            # Import RAGAS components
            from ragas import evaluate
            from ragas import EvaluationDataset
            from ragas.metrics import LLMContextRecall, Faithfulness, FactualCorrectness
            from ragas.llms import LangchainLLMWrapper
            from langchain_openai import ChatOpenAI
            
            print(f"\nPreparing RAGAS evaluation for {system_name}...")
            
            # Check OpenAI API key
            if not os.getenv("OPENAI_API_KEY"):
                print("Warning: OPENAI_API_KEY not found. RAGAS metrics require an OpenAI API key.")
                print("Please set your OPENAI_API_KEY and run the evaluation again.")
                return None
            
            # Setup LLM for evaluation
            evaluator_llm = LangchainLLMWrapper(ChatOpenAI(model="gpt-3.5-turbo"))
            print("Using OpenAI for RAGAS evaluation")
            
            # Process test cases for RAGAS format
            print("\nProcessing test cases for RAGAS format...")
            dataset_entries = []
            
            for idx, row in enumerate(tqdm(test_cases.iterrows(), total=len(test_cases)), 1):
                question = row[1]['question']
                reference_answer = row[1]['reference_answer']
                
                try:
                    # Get contexts and generate response
                    retrieved_contexts = rag_system.get_most_relevant_docs(question)
                    response = rag_system.generate_answer(question, retrieved_contexts)
                    
                    # Add to dataset
                    entry = {
                        "user_input": question,
                        "retrieved_contexts": retrieved_contexts,
                        "response": response,
                        "reference": reference_answer
                    }
                    dataset_entries.append(entry)
                    
                except Exception as e:
                    print(f"Error processing question {idx}: {str(e)}")
            
            # Create RAGAS evaluation dataset
            print("\nCreating RAGAS evaluation dataset...")
            evaluation_dataset = EvaluationDataset.from_list(dataset_entries)
            
            # Save the processed dataset for reference
            dataset_df = pd.DataFrame(dataset_entries)
            dataset_df.to_csv(system_dir / "ragas_dataset.csv", index=False)
            
            # Define metrics
            metrics = [
                LLMContextRecall(),  # How well retrieved documents cover the question
                Faithfulness(),       # Response's factual consistency with context
                FactualCorrectness()  # If response is factually correct vs reference
            ]
            
            # Run RAGAS evaluation
            print(f"\nRunning RAGAS evaluation for {system_name}...")
            result = evaluate(
                dataset=evaluation_dataset,
                metrics=metrics,
                llm=evaluator_llm
            )
            
            # Extract and save metrics
            result_dict = self._extract_ragas_metrics(result)
            
            # Save results
            with open(system_dir / "ragas_results.json", "w") as f:
                json.dump(result_dict, f, indent=4)
            
            # Display results
            print(f"\nRAGAS Evaluation Results for {system_name}:")
            for metric, score in result_dict.items():
                print(f"{metric}: {score:.4f}")
            
            # Create visualization
            import matplotlib.pyplot as plt
            plt.figure(figsize=(10, 6))
            metrics = list(result_dict.keys())
            scores = list(result_dict.values())
            
            # Bar chart
            plt.bar(metrics, scores, color='skyblue')
            plt.ylim(0, 1.0)
            plt.title(f'RAGAS Evaluation Metrics - {system_name}')
            plt.ylabel('Score')
            plt.xticks(rotation=45)
            plt.tight_layout()
            
            # Save visualization
            plt.savefig(system_dir / "ragas_metrics.png")
            plt.close()
            
            print(f"\nRAGAS evaluation complete for {system_name}!")
            print(f"Results saved to {system_dir}")
            
            return result_dict
            
        except Exception as e:
            print(f"Error during RAGAS evaluation: {str(e)}")
            import traceback
            print(traceback.format_exc())
            return None
    
    def _extract_ragas_metrics(self, result) -> Dict[str, float]:
        """Extract metrics from RAGAS evaluation result"""
        result_dict = {}
        
        # Try to extract from pandas DataFrame
        if hasattr(result, 'to_pandas'):
            print("\nExtracting metrics from DataFrame...")
            df = result.to_pandas()
            
            # Print DataFrame columns to understand structure
            print(f"DataFrame columns: {df.columns.tolist()}")
            
            # Get metrics (all non-input columns)
            input_cols = ['user_input', 'response', 'retrieved_contexts', 'reference']
            metric_cols = [col for col in df.columns if col not in input_cols]
            
            # Calculate average for each metric
            for col in metric_cols:
                result_dict[col] = float(df[col].mean())
        
        # Try to extract from __dict__ attribute
        elif hasattr(result, '__dict__'):
            print("\nExtracting metrics from __dict__...")
            for key, value in result.__dict__.items():
                if isinstance(value, (int, float)) and not key.startswith('_'):
                    result_dict[key] = float(value)
        
        # If we still don't have results, check common metric attributes
        if not result_dict:
            print("\nAttempting to extract metrics directly...")
            for attr in ['context_recall', 'faithfulness', 'factual_correctness']:
                if hasattr(result, attr):
                    result_dict[attr] = float(getattr(result, attr))
        
        return result_dict


def display_giskard_report(results_dir):
    """
    Display a previously generated Giskard HTML report in a Jupyter notebook
    
    Args:
        results_dir: Path to the model evaluation results directory (string or Path)
    """
    from IPython.display import HTML, display, Markdown
    from pathlib import Path
    
    # Ensure the path is a Path object
    if isinstance(results_dir, str):
        results_dir = Path(results_dir)
    
    # Display model name as header
    model_name = results_dir.name.replace("_", " ")
    display(Markdown(f"# Giskard Evaluation Report for {model_name}"))
    
    # Path to the Giskard HTML report
    giskard_report_path = results_dir / "giskard_evaluation_report.html"
    
    if giskard_report_path.exists():
        # Read and embed the HTML directly
        with open(giskard_report_path, 'r', encoding='utf-8') as f:
            html_content = f.read()
            
            # Use an iframe for better isolation
            iframe_html = f"""
            <iframe 
                srcdoc="{html_content.replace('"', '&quot;')}" 
                width="100%" 
                height="800px" 
                style="border: 1px solid #ddd; border-radius: 5px;">
            </iframe>
            """
            display(HTML(iframe_html))
        
        print(f"Giskard report displayed from: {giskard_report_path}")
    else:
        display(Markdown("## ⚠️ Giskard report not found"))
        print(f"No Giskard report found at: {giskard_report_path}")
        print("Run the Giskard evaluation first using the run_giskard_evaluation method.")


# display_giskard_report("unified_rag_evaluation/GROQ_llama-3.3-70b-versatile_(temp=0.1)")
```

```python
def evaluate_single_model(model, test_cases, evaluator, run_giskard=True, run_ragas=True):
    """Helper function to evaluate a single model"""
    print(f"\n{'='*50}")
    print(f"Evaluating {model.descriptive_name}")
    print(f"{'='*50}")
    
    # Basic evaluation
    results = evaluator.evaluate_system(
        model, 
        test_cases,
        context_count=4,
        use_parallel=True
    )
    
    # Run Giskard evaluation if requested
    if run_giskard:
        print(f"\nRunning Giskard evaluation for {model.descriptive_name}...")
        evaluator.run_giskard_evaluation(model, test_cases)
    
    # Run RAGAS evaluation if requested
    if run_ragas:
        print(f"\nRunning RAGAS evaluation for {model.descriptive_name}...")
        evaluator.run_ragas_evaluation(model, test_cases)
    
    return results
```
